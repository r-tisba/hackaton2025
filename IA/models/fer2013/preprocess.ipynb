```python
import os
import matplotlib.pyplot as plt
from collections import Counter
```

```python
data_path = "../../data/train"
class_counts = {class_name: len(os.listdir(os.path.join(data_path, class_name))) for class_name in os.listdir(data_path)}

plt.bar(class_counts.keys(), class_counts.values())
plt.xticks(rotation=45)
plt.ylabel("Nombre d'images")
plt.title("Répartition des images par classe dans train")
plt.show()

print(class_counts)

data_path = "../../data/test"
class_counts = {class_name: len(os.listdir(os.path.join(data_path, class_name))) for class_name in os.listdir(data_path)}

plt.bar(class_counts.keys(), class_counts.values())
plt.xticks(rotation=45)
plt.ylabel("Nombre d'images")
plt.title("Répartition des images par classe dans test")
plt.show()

print(class_counts)
```

```python
import os
import cv2
from PIL import Image

# 📂 Dossier contenant les images 32x32
input_folder = "../../data/train/happy"
output_folder = "../../temp/"

# 🔹 Assurer que le dossier de sortie existe
os.makedirs(output_folder, exist_ok=True)

# 🔹 Lister les fichiers image
image_files = [f for f in os.listdir(input_folder) if f.endswith((".jpg", ".png"))][:10]

# 🔄 Redimensionner les images
for img_name in image_files:
    img_path = os.path.join(input_folder, img_name)
    
    # 📷 Charger l'image avec OpenCV
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    
    # 🔄 Redimensionner en 64x64
    img_resized = cv2.resize(img, (64, 64), interpolation=cv2.INTER_AREA)
    
    # 🖼️ Sauvegarder avec PIL pour meilleure compatibilité
    output_path = os.path.join(output_folder, img_name)
    Image.fromarray(img_resized).save(output_path)

    print(f"[✅] {img_name} redimensionnée et sauvegardée -> {output_path}")

print("\n🔹 Redimensionnement terminé ! Vérifie ton dossier 'output_images'.")
```
